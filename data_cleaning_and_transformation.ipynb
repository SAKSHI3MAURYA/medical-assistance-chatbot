```python
data = data.replace({np.nan:0})

data.head(2)
```

```python
ordered_symptoms = []

for i in range(len(data['Disease'])):
    temp_list = []
    for k in range(1,17):
        if data.iloc[i][k] == 0:
            break
        temp_list.append(data.iloc[i][k])
    ordered_symptoms.append(temp_list)

#ordered_symptoms
```

```python
#Using sorting methods to sort all columns by disease
diseases = sorted(data['Disease'].unique())
descs = descriptions.sort_values(by='Disease')
pre_c = precautions.sort_values(by='Disease')
```

```python
#Empty lists to append sorted values to in the precautions data
ordered_cautions = []

#2d sorting method that takes in each precaution and stores it in an individual
#list for the given symptom
for i in range(len(pre_c['Disease'])):
    temp_list = []
    for k in range(1,5):
        temp_list.append(pre_c.iloc[i][k])
    ordered_cautions.append(temp_list)

#Empty lists to append sorted values to in the symptoms data
ordered_symptoms = []

for i in range(len(data['Disease'])):
    temp_list = []
    for k in range(1,17):
        if data.iloc[i][k] == 0:
            break
        temp_list.append(data.iloc[i][k])
    ordered_symptoms.append(temp_list)
```

```python
#Dictionary to hold disease (keys) and its values (a list of the symptoms)
disease_dict = {}

#Looping through the rows
for i in range(len(data['Disease'])):
    #Temp list to hold the symptoms for the disease we're on
    symptoms_list = []
    #Looping through the columns
    for k in range(len(data.columns)):
        #If an item is a 0 or is already in the keys, then ignore
        if data.iloc[i][k] == 0 or data.iloc[i][k] in disease_dict.keys():
            continue
        #Append the values for each row
        symptoms_list.append(data.iloc[i][k])

    #Assigning the symptoms list to that disease
    disease_dict[data['Disease'][i]] = symptoms_list
```

```python
#Sorting the keys so that they can be added to the larger dataframe
sorted_keys = sorted(disease_dict.keys())

#List to add the sorted symptom values
symptoms_list = []

#Looping through the sorted keys
for i in range(len(sorted_keys)):
    #Appending the list containing the symptoms from our disease dictionary
    symptoms_list.append(disease_dict[sorted_keys[i]])
```

```python
#Assembling everything into a single dataframe
#Check the lengths of your lists to see which ones are different
print(len(diseases))
print(len(descs['Description']))
print(len(ordered_cautions))
print(len(symptoms_list))

#Investigate the lists that have different lengths to determine why they differ.
#Once you have identified the root cause, correct the code that generates these lists
#so that they are all the same length.  For example, if `ordered_cautions` is too short,
#you may need to revisit the code in the cell `ipython-input-16-39590c862947`.

# Example fix if ordered_cautions is one element short:
# Assume the last disease is missing precautions, so append an empty list
if len(ordered_cautions) == len(diseases) - 1:
    ordered_cautions.append([])

#Once you have verified that the lists are the same length, proceed to create the DataFrame.
# Make sure the length of descs['Description'] matches the other lists
df = pd.DataFrame({"Diseases":diseases,
                   "Descriptions":descs['Description'].iloc[:len(diseases)],  # Select the appropriate number of descriptions
                   "Precautions":ordered_cautions,
                   "Symptoms":symptoms_list})
```
